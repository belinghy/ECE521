```python
%matplotlib inline
import numpy as np
import tensorflow as tf
with np.load("notMNIST.npz") as data:
    images, labels = data["images"], data["labels"]
from matplotlib import pyplot as plt


labels_temp = []
for label in labels:
    temp = np.zeros(10)
    temp[label[0]] = 1
    labels_temp.append(temp)
labels = np.array(labels_temp)
images = images / 255.0
images = images.reshape(28*28, 18720).T

training_images = images[0:15000]
training_labels = labels[0:15000]
validation_images = images[15000:16000]
validation_labels = labels[15000:16000]
testing_images = images[16000:]
testing_labels = labels[16000:]
```

```python
batch_size = 100
    
X = tf.placeholder(tf.float32, shape=(None, 28 * 28))
Y = tf.placeholder(tf.float32, shape=(None, 10))
#W = tf.Variable(np.random.randn(28*28, 10).astype("float32"), name="weight")
W = tf.Variable(tf.zeros([784, 10]))
#b = tf.Variable(np.random.randn(10).astype("float32"), name="bias")
b = tf.Variable(tf.zeros([10]))
    
#valid_dataset = tf.placeholder(tf.float32, shape=(batch_size, 28 * 28))
#valid_labels = tf.placeholder(tf.float32, shape=(batch_size, 10))
  
logits = tf.matmul(X, W) + b
cross_entropy = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits, Y))
  
optimizer = tf.train.MomentumOptimizer(0.001, 0.1).minimize(cross_entropy)
  
train_prediction = tf.nn.softmax(logits)
#valid_prediction = tf.nn.softmax(tf.matmul(valid_dataset, W) + b)
#test_prediction = tf.nn.softmax(tf.matmul(test_dataset, W) + b)

correct_prediction = tf.equal(tf.argmax(logits,1), tf.argmax(Y,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

```python
sess = tf.InteractiveSession()

init = tf.initialize_all_variables()
sess.run(init)



for epoch in range(1000):
  l = np.arange(15000)
  np.random.shuffle(l)
  for i in xrange(150):
    x_batch = training_images[i * 100: (i + 1) * 100]
    y_batch = training_labels[i * 100: (i + 1) * 100]
    cost_np, _ = sess.run([cross_entropy, optimizer],
                          feed_dict={X: x_batch, Y: y_batch})
    #Display logs per epoch step
  if epoch % 10 == 0:
    cost_train, accuracy_train = sess.run([cross_entropy, accuracy],
                                          feed_dict={X: training_images, Y: training_labels})
    cost_eval, accuracy_eval = sess.run([cross_entropy, accuracy],
                                                   feed_dict={X: validation_images, Y: validation_labels}) 
    errors_train = training_images.shape[0]*(1-accuracy_train)
    errors_eval = validation_images.shape[0]*(1-accuracy_eval)
    print ("Epoch:%04d, t_cost=%0.9f, t_acc=%0.4f, t_err=%0.4f, v_cost=%0.9f, v_acc=%0.4f, v_err=%0.4f" %
           (epoch+1, cost_train, accuracy_train, errors_train, cost_eval, accuracy_eval, errors_eval))
```
